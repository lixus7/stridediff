# Context-FID Score Presentation
## Necessary packages and functions call

- Context-FID score: A useful metric measures how well the the synthetic time series windows ”fit” into the local context of the time series

```python
import os
import torch
import numpy as np
import sys
sys.path.append(os.path.join(os.path.dirname('__file__'), '../'))
from Utils.context_fid import Context_FID
from Utils.metric_utils import display_scores
from Utils.cross_correlation import CrossCorrelLoss
```

```python
#  V1---结果-ETTh

iterations = 5
# ori_data = np.load('../toy_exp/samples/sine_ground_truth_24_train.npy')
ori_data = np.load('../OUTPUT/etth/samples/etth_norm_truth_24_train.npy')  # Uncomment the line if dataset other than Sine is used.
fake_data = np.load('../OUTPUT/etth/ddpm_fake_etth_milestone_10_big30_med20_small1_last12_V2.npy')

for j in range(3):

    context_fid_score = []

    for i in range(iterations):
        context_fid = Context_FID(ori_data[:], fake_data[:ori_data.shape[0]])
        context_fid_score.append(context_fid)
        print(f'Iter {i}: ', 'context-fid =', context_fid, '\n')

    display_scores(context_fid_score)

# Seed 12345 Final Score:  0.13663267402399055 ± 0.005536055372540525
# Seed 12345 Final Score:  0.129
# Seed 12345 Final Score:  0.135

# Seed 2025 Final Score:  0.148

# v1-decay_factor=0.1:   0.5-0.6
# v1-decay_factor=0.2:   0.7-0.8

# v2-升级到 DPM-Solver 2nd-order（二阶多步法）
#  Big30 med20 small1    C-FID  0.25左右。 13.74s
#  Big10 med5 small1    C-FID  0.23左右。 37.55s

def random_choice(size, num_select=100):
    select_idx = np.random.randint(low=0, high=size, size=(num_select,))
    return select_idx
x_real = torch.from_numpy(ori_data)
x_fake = torch.from_numpy(fake_data)

correlational_score = []
size = int(x_real.shape[0] / iterations)

for i in range(iterations):
    real_idx = random_choice(x_real.shape[0], size)
    fake_idx = random_choice(x_fake.shape[0], size)
    corr = CrossCorrelLoss(x_real[real_idx, :, :], name='CrossCorrelLoss')
    loss = corr.compute(x_fake[fake_idx, :, :])
    correlational_score.append(loss.item())
    print(f'Iter {i}: ', 'cross-correlation =', loss.item(), '\n')

display_scores(correlational_score)
```

## ETTh 四组 jump 参数对比 (big_k, med_k, small_k, last_k_always_micro)

对以下 4 组采样结果分别计算 Context-FID 与 cross-correlation，便于对比：

| 配置 | big_k | med_k | small_k | last_k_always_micro |
|------|-------|-------|---------|---------------------|
| 1    | 40    | 30    | 1       | 12                  |
| 2    | 50    | 30    | 1       | 12                  |
| 3    | 20    | 10    | 1       | 12                  |
| 4    | 10    | 5     | 1       | 12                  |

```python
# ETTh 四组 jump 参数结果：加载并计算 Context-FID + cross-correlation
iterations = 5
ori_data = np.load('../OUTPUT/etth/samples/etth_norm_truth_24_train.npy')
base_dir = '../OUTPUT/etth'

# 4 组实验：(标签, 文件名)
etth_configs = [
    ('big_k=40, med_k=30, small_k=1, last=12', 'ddpm_fake_etth_milestone_10_big40_med30_small1_last12.npy'),
    ('big_k=50, med_k=30, small_k=1, last=12', 'ddpm_fake_etth_milestone_10_big50_med30_small1_last12.npy'),
    ('big_k=20, med_k=10, small_k=1, last=12', 'ddpm_fake_etth_milestone_10_big20_med10_small1_last12.npy'),
    ('big_k=10, med_k=5,  small_k=1, last=12', 'ddpm_fake_etth_milestone_10_big10_med5_small1_last12.npy'),
]

def random_choice(size, num_select=100):
    select_idx = np.random.randint(low=0, high=size, size=(num_select,))
    return select_idx

results_etth = []
for label, fname in etth_configs:
    path = os.path.join(base_dir, fname)
    if not os.path.isfile(path):
        print(f'[SKIP] {label}: file not found {path}')
        results_etth.append((label, None, None))
        continue
    fake_data = np.load(path)
    context_fid_score = []
    for i in range(iterations):
        context_fid = Context_FID(ori_data[:], fake_data[:ori_data.shape[0]])
        context_fid_score.append(context_fid)
        print(f'  Iter {i}: context-fid = {context_fid}')
    mean_cf, std_cf = np.mean(context_fid_score), np.std(context_fid_score)
    display_scores(context_fid_score)
    print(f'ETTh ({label}) Seed 12345 Context-FID Final Score: {mean_cf:.4f} ± {std_cf:.4f}\n')

    x_real = torch.from_numpy(ori_data)
    x_fake = torch.from_numpy(fake_data)
    size = int(x_real.shape[0] / iterations)
    correlational_score = []
    for i in range(iterations):
        real_idx = random_choice(x_real.shape[0], size)
        fake_idx = random_choice(x_fake.shape[0], size)
        corr = CrossCorrelLoss(x_real[real_idx, :, :], name='CrossCorrelLoss')
        loss = corr.compute(x_fake[fake_idx, :, :])
        correlational_score.append(loss.item())
    mean_cc, std_cc = np.mean(correlational_score), np.std(correlational_score)
    display_scores(correlational_score)
    print(f'ETTh ({label}) Seed 12345 Cross-Correlation Final Score: {mean_cc:.4f} ± {std_cc:.4f}\n')
    results_etth.append((label, (mean_cf, std_cf), (mean_cc, std_cc)))

# 汇总表
print('=' * 60)
print('ETTh 四组 jump 参数对比汇总 (Seed 12345)')
print('=' * 60)
for label, cf, cc in results_etth:
    if cf is not None:
        print(f'  {label}')
        print(f'    Context-FID:        {cf[0]:.4f} ± {cf[1]:.4f}')
        print(f'    Cross-Correlation:  {cc[0]:.4f} ± {cc[1]:.4f}')
    else:
        print(f'  {label}: (file not found)')

# 与 notebook 中注释风格一致，便于复制到上方 ETTh 结果注释
print('\n# Seed 12345 Final Score (ETTh 四组 jump 参数):')
for label, cf, _ in results_etth:
    if cf is not None:
        print(f'# {label}  Context-FID: {cf[0]:.4f} ± {cf[1]:.4f}')
```
